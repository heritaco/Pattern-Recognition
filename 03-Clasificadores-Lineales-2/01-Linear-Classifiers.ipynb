# **LINEAR DISCRIMINANT FUNCTIONS**

Consider the two-class classification problem and a **linear discriminant functions**, then the respective **decision hypersurface** in the $l$-dimensional feature space is a **hyperplane**, that is
$$H=\{  \textbf{x}\in\mathbb{R}^l \ | \ g( \textbf{x}) = 0 \}$$
and
$$g( \textbf{x}) := \textbf{w}^T  \textbf{x} + w_0$$
where $\textbf{w}=(w_1,w_2\ldots, w_l)^T$ is a nonzero vector known as the **weight vector** (or **normal vector** of $H$) and $w_0$ as the **threshold**.

> The function $g(\textbf{x})$ goes from $\mathbb{R}^l$ to $\mathbb{R}$.

> Hyperplanes are $l-1$-dimensional affine sets, which divide $\mathbb{R}^l$ into two half-spaces.


If $\textbf{x}_1, \textbf{x}_2$ are two points on the decision hyperplane $H$, then the following is valid
\begin{eqnarray}
0 &=& \textbf{w}^T \textbf{x}_1+w_0 &=& \textbf{w}^T \textbf{x}_2+w_0\\
&=& \textbf{w}^T(\textbf{x}_1-\textbf{x}_2)
\end{eqnarray}

Since the difference vector $\textbf{x}_1 - \textbf{x}_2$ obviously lies on the decision hyperplane (for any $\textbf{x}_1$, $\textbf{x}_2$), It follows from the previous equation that the vector $\textbf{w}$ is **orthogonal** to the decision hyperplane.

In the special case that $w_0=0$, the hyperplane passes through
the origin.

## **Distance point-hyperplane**

Given a **hyperplane** $H$ in $\mathbb{R}^l$ and a point $A$ outside the hyperplane with coordinates $\textbf{x}_A\in \mathbb{R}^l$, the distance from $A$ to $H$ is defined as the minimum distance between $\textbf{x}_A$ and a point $\textbf{x}^{'} \in H$, that is
$$d(A,H) = \min_{\textbf{x}^{'}\in H}\|\textbf{x}_A-\textbf{x}^{'}\|$$
Geometrically, this distance is the length of the perpendicular segment from $\textbf{x}_A$ to $H$, following the direction of the normal vector $\textbf{w}$.

The hyperplane $H$ is defined by a **normal vector** $\textbf{w}$, meaning that any displacement perpendicular to $H$ lies in the direction of  $\textbf{w}$.

We can describe any point along the line perpendicular to $H$ passing through $\textbf{x}_A$ as:

$$x(z) = \textbf{x}_A + z \textbf{w}$$

where $z$ is a scalar that determines the position along the line.

The closest point $\textbf{x}^{'}$ on the hyperplane satisfies:

$$\textbf{w}^T \textbf{x}^{'} + w_0 = 0$$

Substituting $\textbf{x}^{'}$ into the hyperplane equation:

$$\textbf{w}^T (\textbf{x}_A + z \textbf{w}) + w_0 = 0$$

Expanding the dot product:

$$\textbf{w}^T \textbf{x}_A + z \| \textbf{w} \|^2 + w_0 = 0$$

Solving for $z$:

$$z = -\frac{\textbf{w}^T \textbf{x}_A + w_0}{\| \textbf{w} \|^2}$$

The **projection point** onto the hyperplane is:

\begin{eqnarray}
\textbf{x}^{'} &=& \textbf{x}_A + z \textbf{w}\\
&=& \textbf{x}_A - \frac{\textbf{w}^T \textbf{x}_A + w_0}{\| \textbf{w} \|^2} \textbf{w}
\end{eqnarray}


The **distance** is given by the norm of the orthogonal displacement:

$$d(A, H) = \| \textbf{x}_A - \textbf{x}^{'} \|$$

Substituting $\textbf{x}^{'}$:

\begin{eqnarray}
d(A, H) &=& \left\| \textbf{x}_A - \left( \textbf{x}_A - \frac{\textbf{w}^T \textbf{x}_A + w_0}{\| \textbf{w} \|^2} \textbf{w} \right) \right\|\\
&=&  |-1|\left\| \frac{\textbf{w}^T \textbf{x}_A + w_0}{\| \textbf{w} \|^2} \textbf{w} \right\|\\
&=& \left| \frac{\textbf{w}^T \textbf{x}_A + w_0}{\| \textbf{w} \|^2}\right| \| \textbf{w}\|\\
&=&  \frac{|\textbf{w}^T \textbf{x}_A + w_0|}{\| \textbf{w} \|^2} \| \textbf{w}\|\\
&=& \frac{|\textbf{w}^T \textbf{x}_A + w_0|}{\| \textbf{w} \|}
\end{eqnarray}

Therefore
$$ d(A,H) = \frac{|g(\textbf{x})|}{\|\textbf{w}\|}$$

Using that formula, the distance from the origin to the hyperplane is
\begin{eqnarray}
d(\textbf{0}, H) &=& \frac{|g(\textbf{0})|}{\|\textbf{w}\|}\\
&=& \frac{|w_0|}{\|\textbf{w}\|}
\end{eqnarray}

**Remarks**

>  $|g(\textbf{x})|$ is proportional to the Euclidean distance of the point $\textbf{x}$ from the decision hyperplane.

> The function $g(\textbf{x})$ takes positive values ​​for $\textbf{x}$ on one side of the hyperplane, and negative values ​​for points on the other side.



## **Direction**

We can kwnon the *direction*¨of the hyperplane, that is, in which half-plane of $\textbf{w}^T \textbf{x} = 0$ it is located in the following way:

Let $\textbf{x}_0$ the intersection point between the line trough the normal vector $\textbf{w}$ and the hyperplane $H$.

Due $\textbf{x}_0$ is in the line, it has the form
$$\textbf{x}_0 = \lambda \textbf{w}$$
with $\lambda$ a real parameter.

But, because $\textbf{x}_0$ is in the hyperplane
$$\textbf{w}^T \textbf{x}_0 + w_0 = 0$$

Combining both expressions we have that
\begin{eqnarray}
\textbf{w}^T \lambda \textbf{w} + w_0 &=& 0\\
\lambda \|\textbf{w}\|^2 + w_0 &=& 0\\
\lambda &=& \frac{-w_0}{\|\textbf{w} \|^2}
\end{eqnarray}

Since the denominator is always positive, then the sign of $\lambda$ is equal to $-\text{sign}(w_0)$

* If $w_0$ is positive, the hyperplane is in the *opposite direction** to $\textbf{w}$.

* If $w_0$ is negative, the hyperplane is in the **same direction** to $\textbf{w}$.

* If $w_0=0$, the hyperplane passes through the origin

# **Visual example**

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
plt.rcParams.update({'font.size': 12})
```

```python
w = np.array([2, 4])
w0 = -8
A = np.array([3, 3])
x = np.linspace(-2, 5, 100)
y = (-w[0]*x -w0) / w[1]
proy = A - ((w @ A + w0) / (w @ w) ) *  w
X1, X2 = np.meshgrid(x, x)
G = w[0] * X1 + w[1] * X2 + w0
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 6), sharey=True, sharex=True)

ax[0].arrow(0, 0, w[0], w[1], length_includes_head=True, head_width=0.1, color="darkblue",  linestyle=':')
ax[0].text(w[0], w[1], r"$w$", color="darkblue")
ax[0].set_title(r"Decision hyperplane and weight vector $w$")

ax[1].arrow(0, 0, A[0], A[1], length_includes_head=True, head_width=0.1, color="red", linestyle=':')
ax[1].plot([A[0], proy[0]], [A[1], proy[1]], 'k--', label="")
ax[1].scatter(proy[0], proy[1], color='black', marker='o')
ax[1].text(A[0], A[1], r"$x_A$", color="red", horizontalalignment="left", va="bottom")
ax[1].text(proy[0], proy[1], r"$x^{'}=Proy_H(x)$", horizontalalignment="right", va="top")
ax[1].text((A[0]+proy[0])/2+.1, (A[1]+proy[1])/2-.1, r"$d(A,H) = \frac{|g(x)|}{\|w\|}$")
ax[1].set_title(r"Distance from the point $x_A$ to the hyperplane $H$")

contour = ax[2].contourf(X1, X2, G, levels=50, cmap='coolwarm', alpha=0.7)
ax[2].contour(X1, X2, G, levels=[0], colors='black', linewidths=2)  # Línea de decisión
ax[2].set_title("Discriminant function $g(x)$")
ax[2].scatter(A[0], A[1], color="red")
ax[2].text(A[0], A[1], f"$g(x_A)={w[0] * A[0] + w[1] * A[1] + w0}$", color="red", horizontalalignment="right", va="bottom")
fig.colorbar(contour, ax=ax[2], label=r"$g(x)$")

for i in range(3):
  ax[i].plot(x, y, 'darkblue',
        label=r"$w^T x + w_0 = 0$")
  ax[i].set_xlabel('x')
  ax[i].set_ylabel('y')
  ax[i].axvline(0, color='gray', linestyle='--')
  ax[i].axhline(0, color='gray', linestyle='--')
  ax[i].legend(loc='upper left')

ax[0].set_aspect("equal")
plt.tight_layout()
plt.show()

plt.show()
```
